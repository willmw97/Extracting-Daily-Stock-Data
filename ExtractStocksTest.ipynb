This an example of extracting stock data from Yahoo finance. The main goal of this project to develop a database that will contain all stocks.

```python
import pandas as pd
import time
import random
import yfinance as yf
import os
import warnings;
from datetime import date
from dateutil.relativedelta import relativedelta
warnings.filterwarnings("ignore");
yf.pdr_override()
```

As an example I test one company and retrieve the last 5 years of stock data.

```python
TICKER = 'AAPL'
begin = '2015-01-01'
curr = '2020-01-01'
```

```python
stock = yf.download(TICKER, start = begin, end = curr)
```

```python
stock
```

For our database we want to mark what company does each company belong to so we append the ticker symbol.

```python
stock['ticker'] = TICKER
```

```python
stock
```

We then create the header names for our database that will start.

```python
column_names = ["Open", "High", "Low", "Close", "Adj Close", "Volume", "Ticker"]

database = pd.DataFrame(columns = column_names)
```

```python
database
```

We load our database file then, which is a list of the stock tickers we want.

```python
df = pd.read_csv('Russell-3000.csv')
```

```python
df
```

We assign the dates we want our stocks to look at.

```python

begin = '2015-01-01'
curr = '2020-01-01'
```

```python
database
```

```python
for ind in df.index:
    try:
        TICKER = df['Ticker'][ind]
        company = yf.download(TICKER, start = begin, end = curr)
        company['Ticker'] = TICKER
        database = database.append(company)
    except ConnectionRefusedError:#this is meant to retry the connection to yahoo finance after we connected already
        time.sleep(30)
        #continue
    except IndexError:
        pass
        #continue
    except:
        continue
```

```python
database
```

```python
database['Date'] = database.index
```

```python
database
```

At this point we can save our database to a csv or directly to a database.

```python
database.to_csv('Russell-3000-Database.csv',index = False)
```
